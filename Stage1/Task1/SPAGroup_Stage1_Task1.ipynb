##Task 1

###Images

```python
logo = cv2.imread("/content/computeronic.jpg",1)
logo_gray = cv2.cvtColor(logo, cv2.COLOR_BGR2GRAY)
```

```python
logo.shape
```

```python
b, g, r = cv2.split(logo)
```

```python
b
```

```python
g
```

```python
r
```

```python
blue_value_for_corners = 39
green_value_for_corners = 22
red_value_for_corners = 1

corner = [blue_value_for_corners, green_value_for_corners, red_value_for_corners]
```

```python
background = cv2.imread("/content/colorful.png",1)
```

```python
background.shape
```

```python
ground = cv2.resize(background, (logo.shape[1], logo.shape[0]), interpolation = cv2.INTER_AREA)
```

```python
ground.shape
```

###Code

```python
_, thresh= cv2.threshold(logo_gray, None, 255, cv2.THRESH_OTSU)
```

```python
plt.imshow(thresh, cmap="gray")
plt.show()
```

```python
my_logo = cv2.bitwise_and(ground, ground, mask = thresh)
```

```python
plt.imshow(my_logo[:,:,::-1])
plt.show()
```

```python
my_logo.shape
```

```python
my_logo[(my_logo[:, :, 0] == 0) & (my_logo[:, :, 1] == 0) & (my_logo[:, :, 2] == 0)] = corner
```

```python
plt.imshow(my_logo[:,:,::-1])
plt.show()
```

```python
cv2.imwrite('Our Logo.jpeg',my_logo)
```
