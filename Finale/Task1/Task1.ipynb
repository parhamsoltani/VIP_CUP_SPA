```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img="second.jpg"


def detect_helmet(image_path):
    # Load the image
    image = cv2.imread(image_path)


    if image is not None:
        # Crop the image to the specified shape
        cropped_image = image[0:250, 250:800]

        # Convert cropped image to grayscale
        gray_image = cv2.cvtColor(cropped_image, cv2.COLOR_BGR2GRAY)

        # Thresholding to detect black color
        _, black_mask = cv2.threshold(gray_image, 45, 255, cv2.THRESH_BINARY_INV)

        # Find contours of the black areas
        contours, _ = cv2.findContours(black_mask, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

        # Assume the largest black area is the helmet
        if contours:
            helmet_contour = max(contours, key=cv2.contourArea)

            # Get bounding rectangle around the detected helmet
            x, y, w, h = cv2.boundingRect(helmet_contour)

            # Check if helmet size is valid
            min_helmet_size = 100
            if w >= min_helmet_size and h >= min_helmet_size:
                cv2.rectangle(cropped_image, (x, y), (x + w, y + h), (0, 255, 0), 2)
                helmet_detected = True
                print("Helmet detected!")
            else:
                print("No helmet detected in the image.")
        else:
            print("No helmet detected in the image.")
    else:
        print(f"Error loading image from path: {image_path}")

    # Display the result
    plt.imshow(cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB))
    plt.title('Helmet Detected or Not?')
    plt.show()

    return helmet_detected

# Example usage
helmet_found = detect_helmet(img)
if helmet_found:
    print("Helmet_detected.")
else:
    print("No helmet detected. Safety first!")
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def detect_greenlight(image_path):
    # Load the image
    image2 = cv2.imread(image_path)
    flag_1 = False

    if image2 is not None:
        # Define the region of interest
        region = image2[400:500, 50:130]

        # Check if the color of the given coordinates is green
        condition_r = np.logical_and(region[:, :, 0] >= 0, region[:, :, 0] < 50)
        condition_g = np.logical_and(region[:, :, 1] >= 200, region[:, :, 1] <= 255)
        condition_b = np.logical_and(region[:, :, 2] >= 0, region[:, :, 2] <= 50)

        if np.all(condition_r) and np.all(condition_g) and np.all(condition_b):
            flag_1 = True

        plt.imshow(cv2.cvtColor(region, cv2.COLOR_BGR2RGB))
        plt.title('Green light Detected or Not?')
        plt.show()
    else:
        print(f"Error loading image from path: {image_path}")

    return flag_1

# Example usage
light_found = detect_greenlight(img)
if light_found:
    print("Green color detected.")
else:
    print("Green color not detected. Safety first!")
```
